```python
#Q2
import matplotlib.pyplot as plt

exp_data = np.random.exponential(scale=1.0, size=100000)
uni_data = np.random.rand(100000)
norm_data = np.random.randn(100000)

plt.figure(figsize=(10,6))
plt.hist(exp_data, density=True, bins=100, histtype="step", color="green", label="Exponential")
plt.hist(uni_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(norm_data, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#Q3
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
ax.set_title("3D Plot of Z=X**2+Y**2")
plt.show()
```

```python
#Q4
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

```python
#Q4
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
#Q4
# Select required features (make sure column names match your CSV exactly)
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_filtered = df[features].dropna()

# Pearson Correlation
pearson_corr = df_filtered.corr(method='pearson')
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation")
plt.show()

# Spearman Correlation
spearman_corr = df_filtered.corr(method='spearman')
sns.heatmap(spearman_corr, annot=True, cmap='viridis')
plt.title("Spearman Correlation")
plt.show()
```

```python
#Q1
import pandas as pd

# Upload CSV file manually in Colab or use path if mounted
df = pd.read_csv('/content/drive/MyDrive/Au_nanoparticle_dataset.csv')  # Adjust path if needed

# Filter only required columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
print("Filtered DataFrame:")
print(new_df.head())
```

```python
#Q2
print("First 20 rows:")
print(new_df.head(20))
```

```python
#Q3
# Mean
mean_vals = new_df.mean()
print("Mean:\n", mean_vals)

# Standard deviation
std_vals = new_df.std()
print("\nStandard Deviation:\n", std_vals)

# Quartiles (25%, 50%, 75%)
quartiles = new_df.quantile([0.25, 0.5, 0.75])
print("\nQuartiles:\n", quartiles)
```

```python
#Q4
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 4, figsize=(20, 4))

features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
for i, feature in enumerate(features):
    axs[i].hist(new_df[feature], bins=30, color='skyblue', edgecolor='black')
    axs[i].set_title(f"{feature} Histogram")
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#Q5
import seaborn as sns

sns.pairplot(new_df)
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

```python
#Q6
g = sns.PairGrid(new_df)

# Bivariate histograms on top-right
g.map_upper(sns.histplot)

#Q6.A
# Histogram + KDE on diagonal
g.map_diag(sns.histplot, kde=True)

#Q6.B
# Bivariate KDE on bottom-left
g.map_lower(sns.kdeplot)

plt.suptitle("PairGrid with KDE and Histograms", y=1.02)
plt.show()
```
